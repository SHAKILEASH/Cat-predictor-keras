```python
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
def load_data():
    train = h5py.File("datasets/train_catvnoncat.h5" , "r")
    test = h5py.File("datasets/test_catvnoncat.h5" , "r")
    train_set_x = np.array(train["train_set_x"][:]) 
    train_set_y = np.array(train["train_set_y"][:])
    test_set_x = np.array(test["test_set_x"][:])
    test_set_y = np.array(test["test_set_y"][:])

    num_classes = np.array(test["list_classes"][:])
    print(train_set_y.shape)
    train_set_y = train_set_y.reshape(1,train_set_y.shape[0])
    test_set_y = test_set_y.reshape(1,test_set_y.shape[0])
    return train_set_x,train_set_y,test_set_x,test_set_y,num_classes 
```

```python
train_X_orgi,train_Y,test_X_orgi,test_Y,classes = load_data()
```

```python
train_X = train_X_orgi/255.0
test_X = test_X_orgi/255.0

print("Train image data ::",train_X.shape)
print("Train label data ::",train_Y.shape)
print("Test image data ::",test_X.shape)
print("Test label data ::",test_Y.shape)
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = (64,64,3)),
    keras.layers.Dense(25, activation = "relu"),
    keras.layers.Dense(5, activation = "relu"),
    keras.layers.Dense(1,activation = "sigmoid")
])
```

```python
model.compile(optimizer = "sgd" , loss = "binary_crossentropy", metrics = ["accuracy"])
model.fit(train_X,train_Y.T,epochs = 300)
```

```python
model.evaluate(test_X,test_Y.T)
```

```python
import cv2
img = cv2.imread("images/catwoman.jpg")
img = cv2.resize(img,(64,64))
img = img.reshape([1,64,64,3])
predict = model.predict_classes(img)
print(predict)
```
